```python
# File path
#filename = '/content/drive/MyDrive/XLab 321/Data/Green shaft - grid, unknown infill% (trial 4)/data.txt'
filename = '/content/drive/MyDrive/Copy of data (2).txt'

#for saving plots later:
last_slash_index = filename.rfind('/')

# Extract the parent directory path
parent_directory = filename[:last_slash_index]

# Read CSV file
df = pd.read_csv(filename)

# Find the index of the first RPM value above 300
start_index = df[df['RPM1'] > 300].index[0]

# Find the index where RPM goes below 200 after the start_index
end_index = df['RPM1'].iloc[start_index:].where(df['RPM1'] < 200).first_valid_index() + start_index

# Trim and downsample
trimmed_df = df.iloc[start_index:end_index]
#downsampled_trimmed_df = trimmed_df
downsampled_trimmed_df = trimmed_df.iloc[::250]  # Take every 250th point

# Convert timestamp to just seconds
downsampled_trimmed_df['Time'] = pd.to_timedelta(downsampled_trimmed_df['Time']).dt.total_seconds()
time_diffs = np.diff(downsampled_trimmed_df['Time'])
indices = np.arange(len(time_diffs))
```

**Plotting**

```python

# plot time differences vs their index
plt.figure(figsize=(8, 6))
plt.plot(indices, time_diffs, marker='o', linestyle='-')
plt.xlabel('Index')
plt.ylabel('Time Differences')
plt.title('Time Differences vs Index')
plt.grid(True)
plt.show()
```

```python

# Plot time vs load
plt.figure()
plt.scatter(downsampled_trimmed_df['Time'], downsampled_trimmed_df["Load"], color='blue', s = 2)  # Set marker size to 10
plt.xlabel('Time (seconds)')
plt.ylabel('Load')
plt.title('Time vs Load (Trimmed)')
plt.show()

print(np.std(downsampled_trimmed_df["Load"]))
print(np.average(downsampled_trimmed_df["Load"]))

print(np.std(downsampled_trimmed_df["RPM1"]))
print(np.average(downsampled_trimmed_df["RPM1"]))

print(np.std(downsampled_trimmed_df["RPM2"]))
print(np.average(downsampled_trimmed_df["RPM2"]))

#%%
# Plot load vs RPM1
plt.figure()
plt.scatter(downsampled_trimmed_df['Load'], downsampled_trimmed_df['RPM1'], color='blue')  # Set marker size to 10
plt.xlabel('Load (Lbs)')
plt.ylabel('RPM1')
plt.xlim(0,0.5)
plt.title('Load vs RPM1 (Trimmed)')
plt.show()
```

```python
# Create the figure and subplots
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Plot RPM1 and RPM2 vs Time
axes[0].plot(downsampled_trimmed_df['Time'], downsampled_trimmed_df['RPM1'], label='RPM1', color='blue')
axes[0].plot(downsampled_trimmed_df['Time'], downsampled_trimmed_df['RPM2'], label='RPM2', color='red')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('RPM')
axes[0].set_title('RPM1 and RPM2 vs Time')
axes[0].legend()

# Plot Time vs Load
axes[1].plot(downsampled_trimmed_df['Time'], downsampled_trimmed_df['Load'], color='green')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Load (Lbs)')
axes[1].set_title('Time vs Load')

plt.tight_layout()
plt.show()

# Create contour plot of Load vs RPM1 and RPM2
fig2, ax2 = plt.subplots(figsize=(8, 6))
contour = ax2.tricontourf(downsampled_trimmed_df['RPM1'], downsampled_trimmed_df['RPM2'], downsampled_trimmed_df['Load'], cmap='viridis')
cbar = plt.colorbar(contour)
cbar.set_label('Load (Lbs)')
ax2.set_xlabel('RPM1')
ax2.set_ylabel('RPM2')
ax2.set_title('Contour Plot: Load vs RPM1 and RPM2')

plt.tight_layout()
plt.show()

# Save the first figure
fig.savefig(parent_directory + "/RPM, Load, Time Graphs")

# Save the second figure
fig2.savefig(parent_directory + "/Contour Plot, Load and Rpms")


```
